`Package`

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy.polynomial import Chebyshev
from matplotlib.font_manager import FontProperties
```

`Let's Go!`

```python
def f(x):
    return 1.0 / ( 1 + 25 * x**2)

n = 11
x1 = np.linspace(-1, 1, n)
x2 = Chebyshev.basis(n).roots()
xd = np.linspace(-1, 1, 200)

c1 = Chebyshev.fit(x1, f(x1), n - 1, domain=[-1, 1]) 
c2 = Chebyshev.fit(x2, f(x2), n - 1, domain=[-1, 1])

print ('插值多項式的最大誤差：')
print ('等距離取樣：', abs(c1(xd) - f(xd)).max())
print ('Chebyshev：', abs(c2(xd) - f(xd)).max())

print ('左圖為使用等距離內差點，右圖為使用Chebyshev內差點\n可看出右圖明顯fit好許多（即龍格現象小）')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(xd, f(xd), '--', label='$f(x)$')
ax1.plot(x1, f(x1), 'o', label='Interpolation Point')
ax1.plot(xd, c1(xd), label='Interpolation Point')
ax1.legend(loc='upper right')

ax2.plot(xd, f(xd), '--', label='$f(x)$')
ax2.plot(x2, f(x2), 'o', label='Interpolation Point')
ax2.plot(xd, c2(xd), label='Interpolation Point')
ax2.legend(loc='upper right')

plt.show()
```
